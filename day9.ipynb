```python
import pandas as pd
import numpy as np

grades = pd.Series([85, None, 92, 45, None, 78, 55])
print("original grades:")
print(grades)

print("\nmissing values:")
print(grades.isnull())

filled_grades = grades.fillna(0)
print("\nfilled grades:")
print(filled_grades)

filtered_grades = filled_grades[filled_grades > 60]
print("\nscores > 60:")
print(filtered_grades)
```
