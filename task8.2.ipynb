```python
import numpy as np

data = np.arange(24)
print("original array:")
print(data)

reshaped_data = data.reshape(4, 3, 2)
print("\nreshaped array (4, 3, 2):")
print(reshaped_data)

transposed_data = np.transpose(reshaped_data, (0, 2, 1))
print("\ntransposed array (4, 2, 3):")
print(transposed_data)
print("\nfinal shape:", transposed_data.shape)
```
